# Identifying Participation Profiles Using K-Means Clustering

While the participation ranking identifies which Olympic-relevant sports demonstrate the strongest youth engagement overall, ranking alone does not fully explain broader participation patterns across sports.

To extend the investigation further, K-Means clustering is used to group sports with similar participation characteristics across youth age groups and gender categories.

Clustering helps identify whether certain sports demonstrate:
- exceptionally strong participation patterns,
- moderate but consistent engagement,
- or comparatively lower youth involvement.

The investigation uses three clusters because the analysis aims to identify broad high-, medium-, and lower-engagement participation profiles across Olympic-relevant sports. Using three clusters improves interpretability while avoiding excessive fragmentation of the participation patterns.

This stage extends the structured analysis beyond simple descriptive ranking and supports deeper strategic interpretation relevant to long-term Olympic audience development and Brisbane 2032 fan engagement planning.

```python
# Select participation variables for clustering
cluster_features = olympic_sports[
    ["5-8", "9-11", "12-14", "Males", "Females"]
]

# Standardise participation values
scaler = StandardScaler()

scaled_features = scaler.fit_transform(
    cluster_features
)

# Apply K-Means clustering
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

# Create cluster labels
olympic_sports["cluster"] = kmeans.fit_predict(
    scaled_features
)

# Preview clustering results
olympic_sports[
    ["Sport", "cluster"]
].sort_values("cluster")
```

# Visualising Olympic Sport Participation Clusters

The clustering results are visualised to better understand how Olympic-relevant sports group together based on youth participation patterns across age groups and gender categories.

Visualising the clusters helps identify which sports demonstrate:

- or comparatively moderate and lower participation profiles.

This provides a clearer strategic understanding of how participation patterns differ across Olympic sports and supports stronger interpretation of potential audience development opportunities leading up to Brisbane 2032.

```python
# Label only major sports to avoid clutter
olympic_sports["label"] = olympic_sports["Sport"].where(
    olympic_sports["youth_engagement_score"] > 200000,
    ""
)

# Create strategic clustering scatter plot
fig = px.scatter(
    olympic_sports,
    x="5-8",
    y="12-14",
    color="cluster",
    size="youth_engagement_score",
    hover_name="Sport",
    text="label",
    title="Olympic Sport Participation Clusters",
    size_max=55
)

# Improve layout
fig.update_layout(
    height=700,
    width=1200,
    title_x=0.5,
    xaxis_title="Participation Ages 5–8",
    yaxis_title="Participation Ages 12–14"
)

# Improve marker and label appearance
fig.update_traces(
    textposition="top center",
    marker=dict(
        line=dict(width=1.5, color="black"),
        opacity=0.85
    )
)

# Add strategic reference lines
fig.add_vline(
    x=100000,
    line_dash="dash",
    line_color="gray"
)

fig.add_hline(
    y=50000,
    line_dash="dash",
    line_color="gray"
)

# Add spacing around the chart
fig.update_xaxes(range=[-20000, 580000])
fig.update_yaxes(range=[-120000, 330000])

fig.show()
```

# Ethical Considerations in Structured Participation Analysis

Several ethical considerations are important when analysing youth sport participation data within the Brisbane 2032 Olympic engagement context.

The AusPlay dataset contains participation information relating to children and early adolescents, meaning that the analysis involves data connected to younger population groups. Although the dataset is aggregated and does not contain personally identifiable information, ethical responsibility is still required when interpreting participation trends involving younger Australians.

One important consideration involves representation bias within participation datasets. Participation rates may not fully capture barriers experienced by all demographic groups, including regional communities, lower-income households, culturally diverse populations, or individuals with limited access to organised sport programs. As a result, participation levels observed within the analysis may not fully represent all Australian youth experiences equally.

Another ethical consideration involves the risk of over-prioritising highly participated sports while unintentionally reducing visibility for smaller Olympic sports. The clustering and participation analysis identified sports such as swimming, football/soccer, and basketball as dominant participation categories. However, focusing exclusively on these sports could contribute to unequal resource allocation and reduced developmental opportunities for niche Olympic sports that may already experience lower visibility and participation support.

The investigation also acknowledges that participation levels alone should not be interpreted as direct indicators of social value, athlete potential, or future Olympic success. Sport participation is influenced by multiple external factors including accessibility, cost, geographic availability, school infrastructure, community investment, and media exposure. These broader contextual influences should be considered when interpreting participation differences across Olympic sports.

Additionally, the clustering analysis simplifies complex participation behaviour into grouped participation profiles. While clustering helps identify broader engagement patterns, it may oversimplify important differences between sports and should therefore be interpreted as a strategic analytical tool rather than an exact representation of all participation experiences.

Overall, the structured participation analysis was conducted with the intention of supporting inclusive and evidence-based Olympic engagement strategies while recognising the broader social, accessibility, and representation factors that may influence youth sport participation across Australia.

## Preparing Guardian API Access

To collect Olympic-related media articles, access to *The Guardian Open Platform API* is required. The API allows article headlines, summaries, and article body text to be retrieved programmatically.

A focused search query is created to collect media coverage associated with Olympic-relevant sports, Australian athletes, and Brisbane 2032 engagement themes.

This stage creates the raw unstructured dataset that will later be transformed into a structured JSON text collection for further analysis.

```python
# Read Guardian API key safely
with open("guardian_key.txt", "r") as file:
    api_key = file.read().strip()

# Create focused Guardian search query
guardian_query = (
    "Brisbane 2032 Olympics Australian athletes youth sport "
    "swimming football basketball gymnastics tennis athletics"
)

# Guardian API endpoint
url = "https://content.guardianapis.com/search"
```

## Collecting Olympic Media Articles from The Guardian API

The Guardian API is used to retrieve multiple pages of sports-related media articles connected to Olympic participation, athlete engagement, and Brisbane 2032 relevance.

To improve analytical depth, article headlines, summaries, and article body text are collected together for each media article. Combining multiple textual components increases the richness of the unstructured dataset and supports stronger TF-IDF and topic modelling analysis later in the investigation.

The collection process also removes unnecessary metadata so that the investigation remains focused specifically on Olympic-related media narratives and youth engagement themes.

```python
# Create empty list to store collected articles
all_articles = []

# Guardian API allows multiple pages of results
for page in range(1, 7):

    params = {
        "api-key": api_key,
        "q": guardian_query,
        "section": "sport",
        "page-size": 50,
        "page": page,
        "show-fields": "headline,trailText,bodyText"
    }

    # Send API request
    response = requests.get(url, params=params)

    # Convert response into JSON format
    data = response.json()

    # Extract article results
    articles = data["response"]["results"]

    # Store relevant article information
    for article in articles:

        all_articles.append({
            "headline": article["fields"].get("headline", ""),
            "summary": article["fields"].get("trailText", ""),
            "body": article["fields"].get("bodyText", "")
        })

# Display number of collected articles
print("Total articles collected:", len(all_articles))
```

## Converting Guardian JSON Data into a Structured Text Dataset

Although the Guardian API returns data in JSON format, the text content must be transformed into a structured dataset before text preprocessing and analytical modelling can be performed.

To support the investigation, article headlines, summaries, and article body text are combined into a single text field for each article. Combining these textual components improves analytical depth because important engagement themes may appear across different sections of an article.

This stage also simplifies later preprocessing, TF-IDF analysis, and topic modelling by creating a consistent text structure across all collected media articles.

```python
# Convert JSON articles into dataframe
guardian_df = pd.DataFrame(guardian_articles)

# Combine text components into one field
guardian_df["combined_text"] = (
    guardian_df["headline"] + " " +
    guardian_df["summary"] + " " +
    guardian_df["body"]
)

# Display combined dataset
guardian_df[
    ["headline", "combined_text"]
].head()
```

## Cleaning and Preparing Media Text for Analysis

Before applying TF-IDF and topic modelling techniques, the collected media text must be cleaned and standardised.

Text preprocessing improves analytical quality by reducing inconsistencies and removing unnecessary textual noise that may distort term frequency calculations. The preprocessing stage converts text into a cleaner and more consistent analytical format suitable for TF-IDF vectorisation and topic modelling.

The cleaning process includes:
- converting text to lowercase,
- removing punctuation,
- removing numerical values,
- removing extra whitespace,
- and standardising the textual structure across all collected articles.

This stage improves the reliability of later pattern discovery and thematic analysis.

```python
# Create text cleaning function
def clean_text(text):

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(r"[^\w\s]", " ", text)

    # Remove numbers
    text = re.sub(r"\d+", " ", text)

    # Remove extra whitespace
    text = re.sub(r"\s+", " ", text)

    return text.strip()

# Apply cleaning function
guardian_df["clean_text"] = guardian_df["combined_text"].apply(clean_text)

# Preview cleaned text
guardian_df[
    ["headline", "clean_text"]
].head()
```

## Auditing the Relevance of the Guardian Corpus

Before applying TF-IDF and topic modelling, a relevance audit is performed across the Guardian article corpus. This checks whether the collected media articles contain key concepts connected to the research question, including Olympic sport, youth engagement, Australian context, Brisbane 2032 relevance, and athlete visibility.

This step is useful because text analysis is only meaningful when the underlying article corpus is relevant to the problem being investigated. The relevance audit therefore helps confirm that the Guardian dataset is suitable for analysing Olympic media engagement themes.

```python
# Create relevance flags based on key research concepts

guardian_df["Youth_Flag"] = guardian_df["clean_text"].str.contains(
    r"\byouth\b|\bchildren\b|\bteenagers\b|\bteen\b|\badolescent\b|\byoung\b",
    case=False,
    na=False
)

guardian_df["Olympic_Flag"] = guardian_df["clean_text"].str.contains(
    r"\bolympic\b|\bolympics\b|\bbrisbane\b|\b2032\b",
    case=False,
    na=False
)

guardian_df["Sport_Flag"] = guardian_df["clean_text"].str.contains(
    r"\bswimming\b|\bfootball\b|\bbasketball\b|\bgymnastics\b|\btennis\b|\bathletics\b|\bathletes\b|\bsport\b",
    case=False,
    na=False
)

guardian_df["Australia_Flag"] = guardian_df["clean_text"].str.contains(
    r"\baustralia\b|\baustralian\b|\bbrisbane\b",
    case=False,
    na=False
)

guardian_df["Engagement_Flag"] = guardian_df["clean_text"].str.contains(
    r"\bfans\b|\baudience\b|\bmedia\b|\bstory\b|\bstories\b|\bvisibility\b|\bidentity\b|\bcommunity\b|\bparticipation\b",
    case=False,
    na=False
)

# Count how many relevance flags each article matches
guardian_df["relevance_score"] = (
    guardian_df["Youth_Flag"].astype(int) +
    guardian_df["Olympic_Flag"].astype(int) +
    guardian_df["Sport_Flag"].astype(int) +
    guardian_df["Australia_Flag"].astype(int) +
    guardian_df["Engagement_Flag"].astype(int)
)

# Article is considered relevant if it matches at least 3 relevance categories
guardian_df["Relevant"] = guardian_df["relevance_score"] >= 3

# Create relevance summary
relevance_summary = pd.DataFrame({
    "Category": [
        "Youth-related articles",
        "Olympic/Brisbane-related articles",
        "Sport-related articles",
        "Australia-related articles",
        "Engagement/media-related articles",
        "Overall relevant articles"
    ],
    "Article_Count": [
        guardian_df["Youth_Flag"].sum(),
        guardian_df["Olympic_Flag"].sum(),
        guardian_df["Sport_Flag"].sum(),
        guardian_df["Australia_Flag"].sum(),
        guardian_df["Engagement_Flag"].sum(),
        guardian_df["Relevant"].sum()
    ]
})

# Calculate relevance percentage
relevance_percentage = guardian_df["Relevant"].mean() * 100

print("Corpus relevance percentage:", round(relevance_percentage, 2), "%")

relevance_summary
```

## Extracting the Most Relevant Guardian Articles

After auditing the Guardian corpus, only the most relevant articles are extracted for advanced text analysis.

Filtering the dataset helps improve the quality of TF-IDF and topic modelling by reducing unrelated media noise and concentrating the investigation on articles most strongly connected to youth participation, Olympic engagement, Australian sport, and Brisbane 2032 narratives.

```python
# Sort articles by relevance score
relevant_articles = guardian_df.sort_values(
    by="relevance_score",
    ascending=False
).copy()

# Keep the top 90% most relevant articles
top_n = int(len(relevant_articles) * 0.90)

relevant_articles = relevant_articles.head(top_n).copy()

# Calculate retained percentage
retained_percentage = (
    len(relevant_articles) / len(guardian_df)
) * 100

print("Relevant articles selected:", len(relevant_articles))
print("Percentage of corpus retained:", round(retained_percentage, 2), "%")

# Preview selected articles
relevant_articles[
    ["headline", "relevance_score"]
].head(10)
```

## Removing Low-Value Textual Noise

Before TF-IDF analysis, additional textual filtering is applied to remove low-value words and sports-news filler language that may distort thematic analysis.

Although standard English stopwords are later removed during TF-IDF vectorisation, this additional preprocessing stage improves analytical quality by removing extremely short terms, repetitive media filler words, and generic reporting language that contributes limited analytical value.

This refinement helps strengthen the reliability of later keyword extraction and topic modelling results.

```python
# Create custom stopword list
custom_stopwords = [
    "said",
    "says",
    "say",
    "will",
    "one",
    "also",
    "new",
    "could",
    "would",
    "year",
    "years",
    "time",
    "people",
    "first",
    "last",
    "made",
    "make",
    "way",
    "day",
    "like",
    "just",
    "told",
    "including"
]

# Remove custom stopwords and short words
def refine_text(text):

    words = text.split()

    filtered_words = [
        word for word in words
        if word not in custom_stopwords
        and len(word) > 2
    ]

    return " ".join(filtered_words)

# Apply refined preprocessing
relevant_articles["refined_text"] = (
    relevant_articles["clean_text"]
    .apply(refine_text)
)

# Preview refined text
relevant_articles[
    ["headline", "refined_text"]
].head()
```

```python
# Create highly focused Olympic engagement corpus
focused_articles = relevant_articles[
    relevant_articles["refined_text"].str.contains(
        r"olympic|brisbane|youth|sport|athlete|australia|games|participation|football|swimming|basketball|tennis|gymnastics|athletics",
        case=False,
        regex=True
    )
].copy()

# Display focused corpus size
print(
    "Focused articles retained:",
    len(focused_articles)
)

# Calculate focused corpus percentage
focused_percentage = (
    len(focused_articles) / len(guardian_df)
) * 100

print(
    "Focused corpus percentage:",
    round(focused_percentage, 2),
    "%"
)

# Preview focused articles
focused_articles[
    ["headline", "relevance_score"]
].head(10)
```

## Applying Count Vectorisation

Before applying TF-IDF weighting, Count Vectorisation is used to examine the frequency distribution of words appearing across the focused Guardian article corpus.

Count Vectorisation converts textual data into numerical frequency representations by counting how often terms appear within the collected documents. This provides an initial understanding of dominant Olympic-related terminology before applying more advanced weighting techniques such as TF-IDF.

The analysis helps identify highly recurring media terms connected to Olympic sport, athlete visibility, competition narratives, and Brisbane 2032 discussions.

```python
# Import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# Create count vectoriser
count_vectoriser = CountVectorizer(
    stop_words="english",
    max_features=1000,
    min_df=3,
    max_df=0.85
)

# Apply count vectorisation
count_matrix = count_vectoriser.fit_transform(
    focused_articles["refined_text"]
)

# Extract feature names
count_features = count_vectoriser.get_feature_names_out()

# Calculate total word frequencies
word_counts = count_matrix.sum(axis=0).A1

# Create frequency dataframe
count_df = pd.DataFrame({
    "Keyword": count_features,
    "Frequency": word_counts
})

# Sort by frequency
count_df = count_df.sort_values(
    by="Frequency",
    ascending=False
)

# Display top frequency terms
count_df.head(15)
```

## Applying TF-IDF Vectorisation

After refining the Guardian media corpus, TF-IDF vectorisation is applied to transform textual content into a numerical representation suitable for pattern discovery and thematic analysis.

TF-IDF (Term Frequency–Inverse Document Frequency) helps identify words that are more important within specific articles while reducing the influence of highly repetitive terms appearing across the entire corpus.

This process strengthens identification of distinctive Olympic engagement themes, athlete narratives, grassroots participation patterns, and media discussion trends connected to Brisbane 2032.

```python
# Create TF-IDF vectoriser
tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=1000,
    min_df=3,
    max_df=0.85
)

# Apply TF-IDF transformation on focused article corpus
tfidf_matrix = tfidf.fit_transform(
    focused_articles["refined_text"]
)

# Display TF-IDF matrix dimensions
print(
    "TF-IDF matrix shape:",
    tfidf_matrix.shape
)

# Extract feature names
feature_names = tfidf.get_feature_names_out()

# Calculate average TF-IDF score for each keyword
tfidf_scores = tfidf_matrix.mean(axis=0).A1

# Create keyword importance dataframe
keyword_df = pd.DataFrame({
    "Keyword": feature_names,
    "TFIDF_Score": tfidf_scores
})

# Sort keywords by importance
keyword_df = keyword_df.sort_values(
    by="TFIDF_Score",
    ascending=False
)

# Improve decimal formatting
pd.options.display.float_format = "{:.6f}".format

# Display top TF-IDF keywords
keyword_df.head(15)
```

## Applying NMF Topic Modelling to the Focused Corpus

After creating a more focused Olympic engagement corpus, NMF topic modelling is applied again to identify clearer and more meaningful media themes.

Using the focused corpus improves topic quality by reducing unrelated sports-news noise and concentrating the model on articles connected to Olympic sport, Australian athletes, youth engagement, and Brisbane 2032 relevance.

This stage helps move the analysis beyond individual keywords by identifying broader media narratives that may support long-term Olympic fan engagement strategies.

```python
# Create NMF topic model
nmf_model = NMF(
    n_components=4,
    random_state=42
)

# Fit NMF model on focused TF-IDF matrix
nmf_model.fit(tfidf_matrix)

# Extract topic keywords
topic_keywords = []

for topic_index, topic in enumerate(nmf_model.components_):

    top_keywords = [
        feature_names[i]
        for i in topic.argsort()[-10:][::-1]
    ]

    topic_keywords.append({
        "Topic": f"Topic {topic_index + 1}",
        "Keywords": ", ".join(top_keywords)
    })

# Create topic summary dataframe
topic_df = pd.DataFrame(topic_keywords)

# Display topic themes
topic_df
```

Topic 2 is primarily connected with football, rugby, and broader team-sport discussions. These themes indicate that highly visible mainstream sports continue to dominate engagement-focused sports media narratives in Australia.

This may influence the analysis by amplifying already dominant sports narratives while underrepresenting less visible sporting groups.

These techniques identify statistical patterns within text data but do not fully capture emotional nuance, social context, sarcasm, cultural meaning, or deeper human interpretation within sporting narratives. As a result, topic modelling outputs should be interpreted cautiously and understood as analytical simplifications rather than exact representations of public opinion or athlete experience.

The analysis additionally recognises the potential influence of algorithmic bias within topic modelling processes. Topic clustering methods may overemphasise repetitive high-frequency terms and dominant narratives while reducing visibility for smaller or emerging themes within the media corpus. This may unintentionally reinforce existing media inequalities surrounding athlete visibility and sport representation.

Privacy and responsible data usage were also considered throughout the investigation. The Guardian articles used within the analysis were publicly available media sources accessed through the official Guardian API. No personal private information or individual-level sensitive data was collected or analysed during the investigation.

Overall, the semi-structured media analysis was conducted with the intention of providing responsible and evidence-based insight into Olympic media engagement patterns while recognising the broader limitations, biases, and ethical challenges associated with automated text analytics and media representation.

# Conclusion

This investigation examined how youth participation patterns and Olympic-related media visibility may contribute to long-term fan engagement strategies leading into the Brisbane 2032 Olympic Games.

The structured AusPlay participation analysis identified several Olympic-relevant sports with consistently strong participation among Australian children and early adolescents, particularly swimming, football/soccer, basketball, athletics, tennis, and gymnastics. These sports demonstrated substantially higher youth engagement levels across multiple age groups and gender categories, suggesting that they already maintain strong grassroots participation foundations among younger Australians. The findings indicate that sports with sustained youth participation may provide particularly valuable opportunities for long-term Olympic audience development because early sporting involvement may strengthen emotional connection, sporting identity, and long-term fan behaviour over time.

The clustering analysis further demonstrated that participation patterns differ significantly across Olympic sports. Three distinct participation profiles emerged within the dataset, separating highly participated mainstream sports from moderate and more niche participation sports. Swimming formed a distinct high-engagement cluster due to its exceptionally strong participation across all analysed participation categories, while football/soccer and basketball also demonstrated broad and sustained youth engagement patterns. These findings suggest that Olympic sports may require different engagement strategies rather than relying on a single broad promotional approach across all sports.

The semi-structured Guardian media analysis extended these findings by demonstrating that many highly participated sports also maintain strong Olympic-related media visibility. The Count Vectorisation analysis identified highly recurring terms connected to sporting competition, athlete performance, Olympic events, and Australian sport culture. Frequently occurring keywords such as *sport*, *athletes*, *games*, *Olympic*, *football*, *tennis*, and *athletics* indicated that mainstream sporting discussions dominate Olympic-related media coverage.

The TF-IDF keyword analysis further identified influential themes connected to athlete achievement, Olympic competition, sporting success, national identity, and Australian sporting culture. Keywords such as *olympic*, *athletes*, *sport*, *athletics*, *football*, and *australia* appeared prominently throughout the analysis, suggesting that athlete visibility and Olympic storytelling may contribute to stronger emotional audience engagement among younger Australians.

The NMF topic modelling analysis additionally identified dominant media narratives focused on athlete performance, Olympic preparation, mainstream sporting discussions, and Australian sporting identity. The Topic Dominance Analysis further demonstrated that a relatively small number of recurring media themes account for a large proportion of the collected Guardian coverage, indicating that Olympic-related media engagement is strongly concentrated around visible athlete narratives and major sporting discussions.

Importantly, integrating both structured participation analysis and semi-structured media analysis provided substantially deeper insight than relying on either dataset independently. Participation analysis identified sports with strong grassroots engagement foundations, while media analysis explained how athlete storytelling, Olympic visibility, recurring sporting themes, and national sporting narratives may reinforce long-term audience connection and fan development opportunities.

The findings additionally suggest that participation behaviour and media visibility may reinforce one another over time. Young people who actively participate in sports may be more likely to follow athletes, engage with Olympic narratives, and maintain long-term sporting interest when those sports also receive sustained media attention and public visibility. This relationship is particularly important within the Brisbane 2032 context because current school-aged participants represent future Olympic audiences, digital consumers, and potentially future athletes themselves.

Overall, the investigation demonstrates that combining structured participation analysis with semi-structured media analysis can provide valuable strategic insight for the Australian Sports Commission. The findings suggest that long-term Brisbane 2032 fan engagement may be strengthened through a combination of youth-focused participation initiatives, athlete-centred media storytelling, school and community engagement programs, and targeted digital visibility strategies that connect younger Australians with Olympic sport, athletes, and national sporting identity.
